```python
n = 12345
rev = 0
while(n!=0):
    rem = n%10
    rev = rev*10+rem
    n//=10
print(rev)
```

```python
n = 12321
rev = 0
while n>rev :
    rem = n%10
    rev = rev*10+rem
    n//=10
if n==rev or n==rev//10:
    print("Palindrome")
else:
    print("Not Palindrome")
```
