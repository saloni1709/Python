```python
# Iterator ek object hai jo elements ko step by step traverse kar sakta hai.
# Iska main feature hai lazy evaluation — matlab ye ek time me sirf current element ko handle karta hai, poora list memory me store nahi karta.
# Python me iterator ke liye do methods compulsory hote hain:
# __iter__() → iterator object return karta hai
# __next__() → next element return karta hai, aur end me StopIteration raise karta hai
```

```python
a = [1, 2, 3, 4]
iter_a = iter(a)
print(next(iter_a))
print(next(iter_a))
print(next(iter_a))
print(next(iter_a))
```
