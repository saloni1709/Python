```python
# Generator ek special type ka iterator hai jo lazy evaluation karta hai.
# Matlab ye ek time me sirf current value produce karta hai, poori sequence memory me store nahi karta.
# Python me generator yield keyword se banta hai.
# Ye functional programming ke lazy evaluation concept se directly related hai.
```

```python
def square(limit):
    x = 0
    while x < limit:
        yield x*x
        yield x*x*x
        x+=1
a = square(5)
print(next(a), next(a))
print(next(a), next(a))
print(next(a), next(a))
```
