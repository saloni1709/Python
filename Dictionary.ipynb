## Dictionary Operations

### 1. Membership Test
##### We can test if a key is in a dict. or not using the keyword 'in'.
##### Membership test is only for the keys & not for the values.

```python
Squares = {
    1 : 1,
    2 : 4,
    3 : 9,
    4 : 16,
    5 : 25
}
print(1 in Squares)
print(3 in Squares)
print(25 in Squares)    # 16 ek value h so ye isko nhi lega and false print hoga it it working only for keys
```

## 2. Iterating through dictionary
#### Iterate (Iteration) ka matlab hota hai 👉 ek-ek element par loop chalana

```python
Squares = {
    1 : 1,
    2 : 4,
    3 : 9,
    4 : 16,
    5 : 25
}
for i in Squares:
    print(Squares[i])
```

```python
Squares = {
    1 : 1,
    2 : 4,
    3 : 9,
    4 : 16,
    5 : 25
}
for i in Squares:
    print(i)
```
