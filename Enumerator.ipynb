enumerate() Python ka built-in function hai

Ye iterable ke saath index + value dono deta hai

Mostly loops me counter maintain karne ke liye use hota hai

# enumerate(iterable, start=0)
iterable → list, tuple, string, etc.

start → index kis number se start karna hai (default = 0)

```python
l1 = ['eat', 'sleep', 'repeat']
for x in enumerate(l1):
      print(x)
# Changing index & printing  separetely
for count, x in enumerate(l1, 100):
         print(count, x)
```
