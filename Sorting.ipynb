```python
l = [10, 3, 23, 7, 20]
x = []
while len(l)>0:
    smallest = l[0]
    for i in range(len(l)):
        if(l[i]<smallest):
            smallest = l[i]
    
    x.append(smallest)
    l.remove(smallest)
print(x)
```
