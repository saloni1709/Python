## Dot product = sum of products of corresponding elements of two equal-length vectors.

```python
# Two vectors x and y
x = [1, 2, 3]
y = [4, 5, 6]

# Step 1: Check if both vectors have same length
# Dot product is only defined for vectors of same size
if len(x) == len(y):
    dot_Product = 0               # Step 2: Initialize dot product to 0

    # Step 3: Loop through each element
    for i in range(len(x)):
        dot_Product += x[i] * y[i]   # Multiply corresponding elements and add to sum

    # Step 4: Print final dot product
    print("Dot Product:", dot_Product)
else:
    # Step 5: If vectors have different lengths, dot product can't be calculated
    print("Dot Product can't be calculated")


## 1∗4+2∗5+3∗6=4+10+18 = 32
```
