## Set Operations

### Set Union (|)

```python
A = {1,2,3}
B = {4,5,6}
print(A|B)
```

```python
A = {1,2,3}
B = {4,5,6,7}
C = A.union(B)
print(C)
```

## Set Intersection (&)

```python
A = {1,4,6}
B = {2,4,6,3}
print(A&B)
```

```python
A = {1,4,6}
B = {2,4,6,3}
C = A.intersection(B)
print(C)
```

## Set Difference (-)

```python
A = {1,4,6}
B = {2,4,6}
print(A-B)
```

```python
A = {1,4,6}
B = {2,4,6}
print(B-A)
```

```python
A = {1,4,6}
B = {2,4,6}
C = A.difference(B)
print(C)
```

## Symmetric Difference (^)
#### Jo common nahi hain (unique in both)

```python
A = {1,4,6}
B = {2,4,6}
print(A^B)
```

```python
A = {1,4,6}
B = {2,4,6}
C = A.symmetric_difference(B)
print(C)
```

## issubset (<=)
#### A ke saare elements B me hain?

```python
A = {1,4,6}
B = {2,4,6,7,9,1}
print(A <= B)
```

```python
A = {1,4,6}
B = {2,4,6,7,9,1}
C = A.issubset(B)
print(C)
```

## issuperset ( >= )
#### B ke saare elements A me hain?

```python
A = {1,4,6}
B = {2,4,6,7,9,1}
print(A>=B)
```

```python
A = {1,4,6}
B = {2,4,6,7,9,1}
C = A.issuperset(B)
print(C)
```

## isdisjoint Sets
#### Koi common element nahi

```python
A = {1,2}
B = {3,4}
C = A.isdisjoint(B)
print(C)
```

```python
A = {1,2}
B = {3,4,1}
C = A.isdisjoint(B)
print(C)
```
