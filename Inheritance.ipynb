## Single Inheritance
#### Jab ek child class, sirf ek parent class se inherit karti hai 
### Parent → Child

```python
class Animal:
    def __init__(self):
        print("Animal class")

class Dog(Animal):
    def __init__(self):
        super().__init__()
        print("Dog class")

d = Dog()
```

## Multiple Inheritance
#### Jab ek child class, ek se zyada parent classes se inherit karti hai.
###      Parent1   Parent2
###             \     /
###             Child


```python
class Animal:
    def __init__(self):
        print("Animal class")

class Pet:
    def __init__(self):
        print("Pet class")

class Dog(Animal, Pet):
    def __init__(self):
        Animal.__init__(self)
        Pet.__init__(self)
        print("Dog class")

d = Dog()       
```

## Multilevel Inheritance
#### Jab ek class, dusri derived class se inherit karti hai
### Grandparent → Parent → Child

```python
class Animal:
    def __init__(self):
        print("Animal class")

class Dog(Animal):
    def __init__(self):
        super().__init__()
        print("Dog class")

class Puppy(Dog):
    def __init__(self):
        super().__init__()
        print("Puppy class")

d = Dog()
p = Puppy()
```

## Hierarchical Inheritance
#### Jab ek hi parent class se multiple child classes inherit karti hain.

```python
    #     Parent
    #    /      \
    # Child1   Child2
```

```python
class Animal:
    def __init__(self):
        print("Animal class")

class Dog(Animal):
    def __init__(self):
        super().__init__()
        print("Dog class")

class Cat(Animal):
    def __init__(self):
        super().__init__()
        print("Cat class")

d = Dog()
c = Cat()
```

## Hybrid Inheritance
#### Jab do ya zyada inheritance types ka combination use hota hai                                                                   (single + multiple + multilevel etc.)

```python
#         Parent
#        /      \
#    Child1   Child2
#         \    /
#          Child3
```

```python
class Animal:
    def __init__(self):
        print("Animal class")

class Mammal(Animal):
    def __init__(self):
        super().__init__()
        print("Mammal class")

class Pet(Animal):
    def __init__(self):
        super().__init__()
        print("Pet Class")

class Dog(Mammal, Pet):
    def __init__(self):
        super().__init__()
        print("Dog class")

d = Dog()
```
