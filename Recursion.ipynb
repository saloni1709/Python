```python
def factorial(x):
    if x == 1:
        return 1
    else:
        return (x*factorial(x-1))
num = 3
print("The factorial of",num, "is : ", factorial(num))
```

## Binary Search
#### Used to search an Element in Sorted Array.
#### Implimented by 2 Methods:-  (i) Iterative Method  (ii) Recursive Method

## 1. Iterative Method

```python
def binary_Search(L, k):
    start = 0                     # starting index
    end = len(L) - 1              # last index

    while start <= end:           # jab tak search space bachi ho
        mid = (start + end) // 2  # middle index

        if L[mid] == k:           # element mil gaya
            return True

        elif L[mid] < k:          # element right side me hoga
            start = mid + 1

        else:                     # element left side me hoga
            end = mid - 1

    return False                  # element nahi mila

L = [3, 1, 7, 5]   # unsorted list
L.sort()           # binary search se pehle sort karna zaroori hai
k = 1

print(binary_Search(L, k))
```
