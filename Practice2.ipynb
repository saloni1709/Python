```python
# Check Anagram
# Anagram ka matlab hota hai: 👉 Do words ke letters same hote hain, bas order change hota hai
s1 = "silent"
s2 = "listen"
if(sorted(s1) == sorted(s2)):
    print("Anagram")
else:
    print("Not Anagram")
```
