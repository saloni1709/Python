```python
# Program to generate random birthdays of 30 people
# and check whether any two people have the same birthday
# (Based on Birthday Paradox concept)

import random          # random numbers generate karne ke liye

l = []                 # empty list banayi birthdays store karne ke liye

# 30 random birthdays generate kar rahe hain
for i in range(30):
    # 1 se 365 ke beech ek random birthday generate karo
    l.append(random.randint(1, 365))

# list ko sort kar rahe hain
# taaki same birthdays paas-paas aa jaayein
l.sort()

# poori birthday list print kar rahe hain (sirf dekhne ke liye)
print(l)

flag = 0               # flag batayega duplicate mila ya nahi

# list ke har element ko next element se compare kar rahe hain
for i in range(len(l) - 1):
    # agar current aur next birthday same ho
    if l[i] == l[i + 1]:
        print("Repeated birthday:", l[i])  # duplicate print karo
        flag = 1
        break          # ek duplicate milte hi loop stop

# agar poore loop mein koi duplicate nahi mila
if flag == 0:
    print("No Repeated birthday")
```
