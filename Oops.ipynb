```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show(self):
        print(self.name, self.age)

s1 = Student("Saloni", 20)
s1.show()
```

```python
class Student:
    def __init__(self, name):
        self.name = name
    
    def show(self):
        print(self.name)

s1 = Student("Saloni")
s1.show()
```

```python
class Student:
    def __init__(self, name):
        self.name = name
    
    def sayhii(self):
        print("Hello, my name is: ", self.name)

s1 = Student("Saloni")
s1.sayhii()
```

```python
class CSStudent:
    stream = 'cse'

    def __init__(self, roll):
        self.roll = roll
    
a = CSStudent(101)
b = CSStudent(102)
print(a.stream)
print(b.stream)
print(a.roll)
```

```python
# Program to show that we can create instance variables inside methods

class CSStudent:
    stream = 'cse'    # Class Variable

    def __init__(self, roll):
        self.roll = roll    # Instance Variable 

    def setAddress(self, address):
        self.address = address   # Instance Variable

    # Retrives instance variable 
    def getAddress(self):
        return self.address
    
# Main Code
a = CSStudent(101)
a.setAddress("Noida, Up")
print(a.getAddress())
```
