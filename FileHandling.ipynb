```python
# Finding a no. form a phn directory file that is very large.

f = open('test2.txt','r')
searching_number = '1234'
for i,line in enumerate(f, start=1):
    parts = line.strip().split(",")
    if len(parts) == searching_number:
        name, number, gender = parts
        if number == searching_number and gender.upper():
            print("Name: {name}, number: {number}")
            break
else:
     print("No. not found")
f.close()
```

```python
# # Genetic Sequences
# Humans have 4 letters for genetic sequence.
# A particular sequence defines if a person has disease or not. Let's open a file and see.

# The genetic sequence we want to search in the file
bp = 'ATGC'  # e.g., could indicate disease marker

# 1️⃣ Open the file in read mode
f = open("test3.txt", 'r')  # 'r' = read mode

# 2️⃣ Read the whole file content as a single string
seq = f.read()  # seq now contains the full DNA sequence from the file

# 3️⃣ Find the first occurrence of bp in seq
# Returns the starting index of the sequence, or -1 if not found
index = seq.find(bp)

# 4️⃣ Check if the sequence was found
if index != -1:
    # If found, print the position (index) in the sequence
    print(f"Sequence found at position {index}")
else:
    # If not found, print a message
    print("Sequence not found!")

# 5️⃣ Close the file
f.close()  # always close the file to free resources
```
