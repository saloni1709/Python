zip() Python ka built-in function hai

Ye multiple iterables (list, tuple, etc.) ko pair (ya tuple) bana kar combine karta hai

Result me ek iterator milta hai

## zip(iterable1, iterable2, ...)


```python
fruits = ['apple', 'mango', 'banana', 'cherry']
size = [5, 5, 6, 6]
print(list(zip(fruits, size)))
```
