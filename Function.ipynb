```python
def EvenOdd(num):
    if num%2==0:
        return "even"
    else:
        return "odd"
print(EvenOdd(2))
print(EvenOdd(7))
```

```python
def absolute_num(num):
    if num>=0:
        return num
    else:
        return -num
    
print(absolute_num(2))
print(absolute_num(4))
print(absolute_num(-1))
```

```python
def myFunction1():
    global x    # global x ko use/modify karne ke liye
    x = x*2
    print("Value of x in function1", x)

def myFunction2():
    global x     # same global x ko access kar rahe hain
    x = x*3
    print("Value of x in function2", x)

x = 2
print("Value of x before function call", x)   # function call se pehle

myFunction1()          # x = 2 * 2 = 4
myFunction2()          # x = 4 * 3 = 12
print("Value of x after function call", x)    # final value
```

## Python Default Argument

```python
def add(a, b = 10):     # a is required parameter, b is default parameter
# Default parameter hamesha last me hota hai
    ans = a+b
    return ans
print(add(2))     # yahan b pass nahi kiya → default b = 10 use hoga
print(add(2, 3))   # yahan b ki value 3 ho jayegi → default overwrite
```

```python
# Function to find minimum in list

def min_list(l):
    m = l[0]
    for i in range(len(l)):
        if l[i] < m:
            m = l[i]
    return m
l = [20, 14, 28, 3, 5]
print(min_list(l))
```

```python
# Function to find maximum in list

def max_list(l):
    m = l[0]
    for i in range(len(l)):
        if l[i] > m:
            m = l[i]
    return m
l = [20, 14, 28, 3, 5]
print(max_list(l))
```

```python
# Function to append a list before another list

# Function banaya jo do lists ko lega
# aur second list (z) ko first list (l) se pehle add karega
def list_append(l, z):

    # newl ek empty list hai
    # isme hum dono lists ke elements store karenge
    newl = []

    # ---- First loop ----
    # yeh loop list z ke har element ko access karega
    for i in range(len(z)):

        # z[i] ka matlab hai z list ka ith element
        # us element ko newl list me add kar rahe hain
        newl.append(z[i])

    # ---- Second loop ----
    # yeh loop list l ke har element ko access karega
    for i in range(len(l)):

        # l[i] ka matlab hai l list ka ith element
        # ab is element ko newl list me add kar rahe hain
        newl.append(l[i])

    # jab dono lists ke elements add ho jaate hain
    # tab final list ko return kar dete hain
    return newl


# l ek list hai (main list)
l = [10, 20, 30]

# z ek list hai (jo pehle add hogi)
z = [40, 50, 60]

# function call karke result print kar rahe hain
print(list_append(l, z))
```

```python
# Function to append a list after given list

def list_append(z, l):
    newl = []
    for i in range(len(l)):
        newl.append(l[i])
    for i in range(len(z)):
        newl.append(z[i])
    return newl
z = [40, 50, 60]
l = [10, 20, 30]
print(list_append(z, l))
```

```python
# Function to calculate average of all list elements

def list_avg(x):
    sum = 0
    for i in range(len(x)):
        sum = sum+x[i]
    avg = sum/len(x)
    return avg
list = [1, 2, 3, 4]
print(list_avg(list))
```

```python
# Sorting using Function

def list_sort(l):
    newl = []
    while len(l)>0:
        m = l[0]
        for i in l:
            if i<m:
                m = i
        newl.append(m)
        l.remove(m)
    return newl
l = [20, 15, 34, 10, 2]
print(list_sort(l))
```

```python
# Function to multiply two matrices A and B

def matrix_mult(A, B):
    # Number of rows and columns of A
    row_A = len(A)
    col_A = len(A[0])
# Rows = 3 → len(A)
# Columns = 3 → len(A[0])
# [0] use kiya kyunki first row me columns ka count same hota hai
# Matlab: len(A[0]) = number of elements in first row → number of column
    
    # Number of rows and columns of B
    row_B = len(B)
    col_B = len(B[0])

    # Check if multiplication is possible
    # Rule: number of columns of A must equal number of rows of B
    if col_A != row_B:
        return "Matrix Multiplication is not possible."
    
    # Initialize result matrix C with zeros
    # Size of C = row_A x col_B
    C = [[0 for _ in range(col_B)] for _ in range(row_A)]       #ex.-> 2X3(A)  3X2(B) = 2X2(C)
    
    # Multiply matrices
    for i in range(row_A):         # Loop over rows of A   # row of A → C ke rows
        for j in range(col_B):     # Loop over columns of B    # column of B → C ke columns
            for k in range(col_A): # Loop over elements of the row of A / column of B     # sum over A row / B column
                # Multiply corresponding elements and add to C[i][j]
                C[i][j] += A[i][k] * B[k][j]
    
    return C  # Return the result matrix

# Example matrices
A = [
    [1, 2, 3],
    [4, 5, 6],
    [0, 1, 2]
]

B = [
    [7, 8, 9],
    [10, 11, 12],
    [0, 4, 5]
]

# Function call
result = matrix_mult(A, B)

# Print the result
print(result)
```
