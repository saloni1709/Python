```python
# Inline statement basically ek single line me likha gaya statement hota hai.
# Python me ek hi line me multiple expressions ya statements bhi likhe ja sakte hain, ; se separate karke.
# Mostly simple operations ke liye use hota hai, readability ke liye avoid karna chahiye agar complex ho.
```

### Inline If
#### It is a concise version of if...else statement that can be written in just one line.

```python
a = True
if a: print('True')
```

### Inline If with else Statement

```python
color = 'blue'
a = 'Color is red' if color == 'red' else 'Blue'
print(a)
```

```python
d = 4
print('Greater' if d>3 else 'Smaller')
```

### Inline if...elif

```python
value = 10
print('Less than 10' if value < 10 else 'More than 10' if value > 10 else 'Equal to 10')
```

### Inline While Loop

```python
a = 5
while a>0: print(a); a-=1
```
