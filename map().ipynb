map() Python ka built-in functional programming function hai

Ye ek function ko iterable ke har element par apply karta hai

Result me map object (iterator) return hota hai

## map(function, iterable)
function → jo har element par apply hoga

iterable → list, tuple, string, etc.

```python
def addition(n):
    return n+n

# We double all numbers using map()
numbers = (1,2,3,4)
results = map(addition, numbers)
print(results)   # does not print values

# For printing value
for result in results:
    print(result, end=' ')
```

```python
a = [10, 20, 30, 40, 50]
b = [1, 2, 3, 4, 5]
def sub(x, y):
    return x-y
c = map(sub, a, b)
print(list(c))
```

## Filter()

filter() Python ka built-in functional programming function hai

Ye iterable se sirf un elements ko select karta hai jo condition (True/False) satisfy karte hain

Result me filter object (iterator) return hota hai

## filter(function, iterable)
function → True / False return karta hai

iterable → list, tuple, string, etc.

```python
# Function that filter vowels
def func(variable):
    letters = ['a', 'e', 'i', 'o', 'u']
    if(variable in letters):
        return True
    else:
        return False
sequence = ['g', 'e', 'e', 'j', 'k', 's', 'p', 'r']
filtered = filter(func, sequence)
print('The filtered letters are:')
for s in filtered:
    print(s)
```
