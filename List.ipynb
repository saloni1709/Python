# Aliasing of List

### Aliasing means two or more variables refer to the SAME list in memory.

```python
a = [1,2,3]
b = a
print(b is a)
```

```python
a = [1,2,3]
b = a.copy()
print(b is a)
```

```python
a = [1,2,3]
b = a
b[0] = 10
print(a)
print(b)
```

## Compairing Two Lists

```python
l1 = [1,2,3]
l2 = [1,2,3]
l3 = [1,3,2]
l4 = [1,2]
print(l1==l2)
print(l1==l3)
print(l1==l4)
print("\n")
print(l1<l2)
print(l1<l3)
print(l1>l3)
print(l1<l4)
print(l1>l4)
```
