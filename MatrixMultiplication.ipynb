```python
r1 = [1,2,3]
r2 = [4,5,6]
r3 = [7,8,9]

s1 = [1,2,1]
s2 = [6,2,3]
s3 = [4,2,1]

A = []
B = []
A.append(r1)
A.append(r2)
A.append(r3)
B.append(s1)
B.append(s2)
B.append(s3)

c = [[0,0,0], [0,0,0], [0,0,0]]
dim = 3

for i in range(dim):
    for j in range(dim):
        for k in range(dim):
            c[i][j]+=A[i][k]*B[k][j]

print(c)
```

```python
# Matrix A (3x3)
A = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

# Matrix B (3x3)
B = [
    [9,8,7],
    [6,5,4],
    [3,2,1]
]

# Result matrix C initialized to zero (3x3)
C = [
    [0,0,0],
    [0,0,0],
    [0,0,0]
]

# Dimension of matrices (3x3)
dim = 3

# Nested loops to calculate matrix multiplication
for i in range(dim):        # Loop over rows of A
    for j in range(dim):    # Loop over columns of B
        for k in range(dim):  # Loop over summation index (columns of A / rows of B)
            # Summation matlab: sab multiply karke add karna
            # Multiply element of A row by element of B column and add to C[i][j]
            C[i][j] += A[i][k] * B[k][j]

# Print the resulting matrix C
print(C)
```
